```python
import pandas as pd

# Load original dataset
df = pd.read_csv("adulteration_dataset.csv")

spectral_cols = df.columns[4:-1]

# Create 3 demo samples
sample1 = df[df['Concentration_Class'] == 0][spectral_cols].iloc[[0]]
sample2 = df[df['Concentration_Class'] == 25][spectral_cols].iloc[[0]]
sample3 = df[df['Concentration_Class'] == 50][spectral_cols].iloc[[0]]

sample1.to_csv("demo_input_0_percent.csv", index=False)
sample2.to_csv("demo_input_25_percent.csv", index=False)
sample3.to_csv("demo_input_50_percent.csv", index=False)

print("Demo files created successfully!")
```

```python
demo_multi = df.sample(5)[spectral_cols]
demo_multi.to_csv("demo_multiple_samples.csv", index=False)

print("Multi-sample demo file created!")
```
